```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import time

# --- 1. Define the Custom Layer ---
class ReLURelUPlusC(layers.Layer):
    """
    Custom Keras layer implementing ReLU(ReLU(wx+b) + c).
    """
    def __init__(self, units, **kwargs):
        super(ReLURelUPlusC, self).__init__(**kwargs)
        self.units = units
        self.dense_layer = layers.Dense(units) # For wx+b
        # 'c' will be a learnable parameter, initialized to zeros or random
        self.c = self.add_weight(
            name='c',
            shape=(self.units,),
            initializer='zeros', # You can try other initializers
            trainable=True
        )

    def call(self, inputs):
        x = self.dense_layer(inputs) # wx+b
        x = tf.nn.relu(x)           # ReLU(wx+b)
        x = x + self.c              # ReLU(wx+b) + c
        x = tf.nn.relu(x)           # ReLU(ReLU(wx+b) + c)
        return x

    def get_config(self):
        config = super(ReLURelUPlusC, self).get_config()
        config.update({"units": self.units})
        return config

# --- 2. Load and Preprocess Benchmark Datasets ---
def load_and_preprocess_mnist():
    """Loads and preprocesses the MNIST dataset."""
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    # Normalize pixel values to be between 0 and 1
    x_train = x_train.astype("float32") / 255.0
    x_test = x_test.astype("float32") / 255.0

    # Flatten the images
    x_train = x_train.reshape(-1, 28 * 28)
    x_test = x_test.reshape(-1, 28 * 28)

    # One-hot encode labels
    y_train = keras.utils.to_categorical(y_train, num_classes=10)
    y_test = keras.utils.to_categorical(y_test, num_classes=10)

    return (x_train, y_train), (x_test, y_test)

def load_and_preprocess_cifar10():
    """Loads and preprocesses the CIFAR-10 dataset."""
    (x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

    # Normalize pixel values to be between 0 and 1
    x_train = x_train.astype("float32") / 255.0
    x_test = x_test.astype("float32") / 255.0

    # Flatten the images (CIFAR-10 images are 32x32x3)
    x_train = x_train.reshape(-1, 32 * 32 * 3)
    x_test = x_test.reshape(-1, 32 * 32 * 3)

    # One-hot encode labels
    y_train = keras.utils.to_categorical(y_train, num_classes=10)
    y_test = keras.utils.to_categorical(y_test, num_classes=10)

    return (x_train, y_train), (x_test, y_test)

# --- 3. Define Baseline Models (Standard Dense Layers) ---
def create_baseline_model(input_shape, num_classes, hidden_units=[128, 64]):
    """Creates a standard MLP model with Dense layers."""
    model = models.Sequential(name="Baseline_MLP")
    model.add(layers.InputLayer(input_shape=input_shape))
    for units in hidden_units:
        model.add(layers.Dense(units, activation='relu'))
        model.add(layers.Dropout(0.2)) # Optional: Add dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax'))
    return model

# --- 4. Define Models with Your Custom Layer ---
def create_custom_model(input_shape, num_classes, hidden_units=[128, 64]):
    """Creates an MLP model with the custom ReLURelUPlusC layer."""
    model = models.Sequential(name="Custom_MLP_ReLURelUPlusC")
    model.add(layers.InputLayer(input_shape=input_shape))
    for units in hidden_units:
        model.add(ReLURelUPlusC(units))
        model.add(layers.Dropout(0.2)) # Optional: Add dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax')) # Output layer remains standard
    return model

# --- 5. Set Up Training and Evaluation ---
def compile_train_evaluate(model, x_train, y_train, x_test, y_test, epochs=10, batch_size=32):
    """Compiles, trains, and evaluates a given model."""
    print(f"\n--- Training and Evaluating: {model.name} ---")
    start_time = time.time()

    model.compile(optimizer=keras.optimizers.Adam(),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        validation_split=0.1, # Use part of training data for validation
                        verbose=1) # Set to 0 for less output, 1 for progress bar

    end_time = time.time()
    training_time = end_time - start_time
    print(f"Training completed in {training_time:.2f} seconds.")

    print("\nEvaluating on test data:")
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Loss: {loss:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")

    return history, loss, accuracy, training_time

# --- 6. Run Experiments and Compare Performance ---
def run_experiments():
    """Runs experiments on selected datasets and models."""
    datasets = {
        "MNIST": load_and_preprocess_mnist(),
        "CIFAR-10": load_and_preprocess_cifar10()
    }

    results = {}

    # Common training parameters (can be adjusted)
    epochs = 20 # Increased epochs for potentially better convergence
    batch_size = 128
    hidden_units_config = [
        ([64], "Shallow"),          # Shallow network
        ([128, 64], "Medium"),      # Medium depth
        ([256, 128, 64], "Deep")    # Deeper network
    ]


    for dataset_name, ((x_train, y_train), (x_test, y_test)) in datasets.items():
        print(f"\n===================================================")
        print(f"             Benchmarking on {dataset_name}             ")
        print(f"===================================================")
        input_shape = x_train.shape[1:]
        num_classes = y_train.shape[1]
        results[dataset_name] = {}

        for hidden_units, config_name in hidden_units_config:
            print(f"\n--- Configuration: {config_name} ({hidden_units}) ---")
            results[dataset_name][config_name] = {}

            # Baseline Model
            baseline_model = create_baseline_model(input_shape, num_classes, hidden_units)
            baseline_model.summary()
            history_baseline, loss_baseline, acc_baseline, time_baseline = compile_train_evaluate(
                baseline_model, x_train, y_train, x_test, y_test, epochs, batch_size
            )
            results[dataset_name][config_name]["Baseline_MLP"] = {
                "history": history_baseline,
                "test_loss": loss_baseline,
                "test_accuracy": acc_baseline,
                "training_time": time_baseline,
                "num_params": baseline_model.count_params()
            }

            # Custom Model
            custom_model = create_custom_model(input_shape, num_classes, hidden_units)
            custom_model.summary() # To see the structure and parameters
            history_custom, loss_custom, acc_custom, time_custom = compile_train_evaluate(
                custom_model, x_train, y_train, x_test, y_test, epochs, batch_size
            )
            results[dataset_name][config_name]["Custom_MLP_ReLURelUPlusC"] = {
                "history": history_custom,
                "test_loss": loss_custom,
                "test_accuracy": acc_custom,
                "training_time": time_custom,
                "num_params": custom_model.count_params()
            }

            # Optional: Plot training history for this configuration
            plot_training_history(
                history_baseline,
                history_custom,
                f"{dataset_name} - {config_name}: Training History",
                baseline_model.name,
                custom_model.name
            )

    # --- Display final results ---
    print("\n\n===================================================")
    print("                 Overall Results Summary             ")
    print("===================================================")
    for dataset_name, dataset_results in results.items():
        print(f"\n--- {dataset_name} ---")
        for config_name, config_results in dataset_results.items():
            print(f"  -- Configuration: {config_name} --")
            for model_name, metrics in config_results.items():
                print(f"    Model: {model_name}")
                print(f"      Test Accuracy: {metrics['test_accuracy']:.4f}")
                print(f"      Test Loss:     {metrics['test_loss']:.4f}")
                print(f"      Train Time:    {metrics['training_time']:.2f}s")
                print(f"      Num Params:    {metrics['num_params']}")
        print("-" * 50)

    return results

def plot_training_history(history1, history2, title, model1_name="Model 1", model2_name="Model 2"):
    """Plots training accuracy and loss for two models."""
    plt.figure(figsize=(12, 4))

    # Plot Training Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history1.history['accuracy'], label=f'{model1_name} Train Acc')
    plt.plot(history1.history['val_accuracy'], label=f'{model1_name} Val Acc')
    plt.plot(history2.history['accuracy'], label=f'{model2_name} Train Acc', linestyle='--')
    plt.plot(history2.history['val_accuracy'], label=f'{model2_name} Val Acc', linestyle='--')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')

    # Plot Training Loss
    plt.subplot(1, 2, 2)
    plt.plot(history1.history['loss'], label=f'{model1_name} Train Loss')
    plt.plot(history1.history['val_loss'], label=f'{model1_name} Val Loss')
    plt.plot(history2.history['loss'], label=f'{model2_name} Train Loss', linestyle='--')
    plt.plot(history2.history['val_loss'], label=f'{model2_name} Val Loss', linestyle='--')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')

    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make space for suptitle
    plt.show()


if __name__ == '__main__':
    # Ensure TensorFlow is using the GPU if available
    physical_devices = tf.config.list_physical_devices('GPU')
    if physical_devices:
        try:
            for device in physical_devices:
                tf.config.experimental.set_memory_growth(device, True)
            print(f"Using GPU: {physical_devices[0]}")
        except RuntimeError as e:
            # Memory growth must be set before GPUs have been initialized
            print(e)
    else:
        print("No GPU found, using CPU.")

    # Run the experiments
    experiment_results = run_experiments()

    # You can now use `experiment_results` to analyze data for your paper.
    # For example, print specific results:
    # print(experiment_results['MNIST']['Medium']['Custom_MLP_ReLURelUPlusC']['test_accuracy'])
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import numpy as np
import matplotlib.pyplot as plt
import time

# --- 1. Define the NEW Custom Layer ---

class CustomGatedActivationLayer(layers.Layer):
    """
    Implements the function: Step(W1x + b1) * ReLU(W2x + b2)

    This layer has two internal dense transformations:
    1. Gate path: W1x + b1, followed by a Step function (or its soft approximation).
    2. Value path: W2x + b2, followed by a ReLU function.

    The outputs of these two paths are then multiplied element-wise.

    Args:
        units (int): Dimensionality of the output space (number of neurons).
        use_soft_gate (bool): If True, uses a sigmoid function (soft gate)
            instead of a hard step function for the gate. This can help with
            training W1 and b1 due to smoother gradients. Defaults to False.
        sigmoid_k (float): Steepness parameter for the sigmoid gate if
            use_soft_gate is True. Higher values make the sigmoid approximate
            the step function more closely. Defaults to 10.0.
        **kwargs: Standard Keras layer keyword arguments.
    """
    def __init__(self, units, use_soft_gate=False, sigmoid_k=10.0, **kwargs):
        super(CustomGatedActivationLayer, self).__init__(**kwargs)
        self.units = units
        self.use_soft_gate = use_soft_gate
        self.sigmoid_k = float(sigmoid_k) # Ensure it's float

        # Weights will be initialized in the build method
        self.W1 = None
        self.b1 = None
        self.W2 = None
        self.b2 = None

    def build(self, input_shape):
        if not isinstance(input_shape, tf.TensorShape):
            input_shape = tf.TensorShape(input_shape)

        last_dim = input_shape[-1]
        if last_dim is None:
            raise ValueError(
                "The last dimension of the input_shape to CustomGatedActivationLayer "
                "must be defined. Found None. Received: input_shape={}".format(input_shape)
            )

        # Parameters for the gate path: W1, b1
        self.W1 = self.add_weight(
            name='W1_gate',
            shape=(last_dim, self.units),
            initializer='glorot_uniform', # A common good default
            trainable=True
        )
        self.b1 = self.add_weight(
            name='b1_gate',
            shape=(self.units,),
            initializer='zeros',
            trainable=True
        )

        # Parameters for the value (ReLU) path: W2, b2
        self.W2 = self.add_weight(
            name='W2_value',
            shape=(last_dim, self.units),
            initializer='glorot_uniform',
            trainable=True
        )
        self.b2 = self.add_weight(
            name='b2_value',
            shape=(self.units,),
            initializer='zeros',
            trainable=True
        )
        super().build(input_shape) # Finalizes the build process, do not call earlier

    def call(self, inputs):
        # Gate computation
        # z1 = W1x + b1
        z1 = tf.matmul(inputs, self.W1) + self.b1

        if self.use_soft_gate:
            # Soft gate using sigmoid: sigma(k * z1)
            gate = tf.sigmoid(self.sigmoid_k * z1)
        else:
            # Hard step function: H(z1) = 1 if z1 >= 0, else 0
            # Note: When using the hard step function (use_soft_gate=False),
            # the gradients for W1 and b1 might become zero for inputs not
            # exactly at the threshold. This can hinder their training.
            # The soft gate (sigmoid) is generally recommended for trainability.
            gate = tf.cast(z1 >= 0, dtype=inputs.dtype)

        # Value computation
        # z2 = W2x + b2
        z2 = tf.matmul(inputs, self.W2) + self.b2
        value = tf.nn.relu(z2) # ReLU(z2)

        # Element-wise multiplication of gate and value
        output = gate * value
        return output

    def get_config(self):
        config = super(CustomGatedActivationLayer, self).get_config()
        config.update({
            "units": self.units,
            "use_soft_gate": self.use_soft_gate,
            "sigmoid_k": self.sigmoid_k,
        })
        return config

# --- 2. Load and Preprocess Benchmark Datasets ---
def load_and_preprocess_mnist():
    """Loads and preprocesses the MNIST dataset."""
    (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

    # Normalize pixel values to be between 0 and 1
    x_train = x_train.astype("float32") / 255.0
    x_test = x_test.astype("float32") / 255.0

    # Flatten the images
    x_train = x_train.reshape(-1, 28 * 28)
    x_test = x_test.reshape(-1, 28 * 28)

    # One-hot encode labels
    y_train = keras.utils.to_categorical(y_train, num_classes=10)
    y_test = keras.utils.to_categorical(y_test, num_classes=10)

    return (x_train, y_train), (x_test, y_test)

def load_and_preprocess_cifar10():
    """Loads and preprocesses the CIFAR-10 dataset."""
    (x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()

    # Normalize pixel values to be between 0 and 1
    x_train = x_train.astype("float32") / 255.0
    x_test = x_test.astype("float32") / 255.0

    # Flatten the images (CIFAR-10 images are 32x32x3)
    x_train = x_train.reshape(-1, 32 * 32 * 3)
    x_test = x_test.reshape(-1, 32 * 32 * 3)

    # One-hot encode labels
    y_train = keras.utils.to_categorical(y_train, num_classes=10)
    y_test = keras.utils.to_categorical(y_test, num_classes=10)

    return (x_train, y_train), (x_test, y_test)

# --- 3. Define Baseline Models (Standard Dense Layers) ---
def create_baseline_model(input_shape, num_classes, hidden_units=[128, 64]):
    """Creates a standard MLP model with Dense layers."""
    model = models.Sequential(name="Baseline_MLP")
    model.add(layers.InputLayer(input_shape=input_shape))
    for units in hidden_units:
        model.add(layers.Dense(units, activation='relu'))
        model.add(layers.Dropout(0.2)) # Optional: Add dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax'))
    return model

# --- 4. Define Models with Your Custom Layer ---
def create_custom_model(input_shape, num_classes, hidden_units=[128, 64]):
    """Creates an MLP model with the custom CustomGatedActivationLayer layer."""
    model = models.Sequential(name="Custom_MLP_CustomGatedActivationLayer")
    model.add(layers.InputLayer(input_shape=input_shape))
    for units in hidden_units:
        model.add(CustomGatedActivationLayer(units))
        model.add(layers.Dropout(0.2)) # Optional: Add dropout for regularization
    model.add(layers.Dense(num_classes, activation='softmax')) # Output layer remains standard
    return model

# --- 5. Set Up Training and Evaluation ---
def compile_train_evaluate(model, x_train, y_train, x_test, y_test, epochs=10, batch_size=32):
    """Compiles, trains, and evaluates a given model."""
    print(f"\n--- Training and Evaluating: {model.name} ---")
    start_time = time.time()

    model.compile(optimizer=keras.optimizers.Adam(),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    history = model.fit(x_train, y_train,
                        batch_size=batch_size,
                        epochs=epochs,
                        validation_split=0.1, # Use part of training data for validation
                        verbose=1) # Set to 0 for less output, 1 for progress bar

    end_time = time.time()
    training_time = end_time - start_time
    print(f"Training completed in {training_time:.2f} seconds.")

    print("\nEvaluating on test data:")
    loss, accuracy = model.evaluate(x_test, y_test, verbose=0)
    print(f"Test Loss: {loss:.4f}")
    print(f"Test Accuracy: {accuracy:.4f}")

    return history, loss, accuracy, training_time

# --- 6. Run Experiments and Compare Performance ---
def run_experiments():
    """Runs experiments on selected datasets and models."""
    datasets = {
        "MNIST": load_and_preprocess_mnist(),
        "CIFAR-10": load_and_preprocess_cifar10()
    }

    results = {}

    # Common training parameters (can be adjusted)
    epochs = 20 # Increased epochs for potentially better convergence
    batch_size = 128
    hidden_units_config = [
        ([64], "Shallow"),          # Shallow network
        ([128, 64], "Medium"),      # Medium depth
        ([256, 128, 64], "Deep")    # Deeper network
    ]


    for dataset_name, ((x_train, y_train), (x_test, y_test)) in datasets.items():
        print(f"\n===================================================")
        print(f"             Benchmarking on {dataset_name}             ")
        print(f"===================================================")
        input_shape = x_train.shape[1:]
        num_classes = y_train.shape[1]
        results[dataset_name] = {}

        for hidden_units, config_name in hidden_units_config:
            print(f"\n--- Configuration: {config_name} ({hidden_units}) ---")
            results[dataset_name][config_name] = {}

            # Baseline Model
            baseline_model = create_baseline_model(input_shape, num_classes, hidden_units)
            baseline_model.summary()
            history_baseline, loss_baseline, acc_baseline, time_baseline = compile_train_evaluate(
                baseline_model, x_train, y_train, x_test, y_test, epochs, batch_size
            )
            results[dataset_name][config_name]["Baseline_MLP"] = {
                "history": history_baseline,
                "test_loss": loss_baseline,
                "test_accuracy": acc_baseline,
                "training_time": time_baseline,
                "num_params": baseline_model.count_params()
            }

            # Custom Model
            custom_model = create_custom_model(input_shape, num_classes, hidden_units)
            custom_model.summary() # To see the structure and parameters
            history_custom, loss_custom, acc_custom, time_custom = compile_train_evaluate(
                custom_model, x_train, y_train, x_test, y_test, epochs, batch_size
            )
            results[dataset_name][config_name]["Custom_MLP_CustomGatedActivationLayer"] = {
                "history": history_custom,
                "test_loss": loss_custom,
                "test_accuracy": acc_custom,
                "training_time": time_custom,
                "num_params": custom_model.count_params()
            }

            # Optional: Plot training history for this configuration
            plot_training_history(
                history_baseline,
                history_custom,
                f"{dataset_name} - {config_name}: Training History",
                baseline_model.name,
                custom_model.name
            )

    # --- Display final results ---
    print("\n\n===================================================")
    print("                 Overall Results Summary             ")
    print("===================================================")
    for dataset_name, dataset_results in results.items():
        print(f"\n--- {dataset_name} ---")
        for config_name, config_results in dataset_results.items():
            print(f"  -- Configuration: {config_name} --")
            for model_name, metrics in config_results.items():
                print(f"    Model: {model_name}")
                print(f"      Test Accuracy: {metrics['test_accuracy']:.4f}")
                print(f"      Test Loss:     {metrics['test_loss']:.4f}")
                print(f"      Train Time:    {metrics['training_time']:.2f}s")
                print(f"      Num Params:    {metrics['num_params']}")
        print("-" * 50)

    return results

def plot_training_history(history1, history2, title, model1_name="Model 1", model2_name="Model 2"):
    """Plots training accuracy and loss for two models."""
    plt.figure(figsize=(12, 4))

    # Plot Training Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history1.history['accuracy'], label=f'{model1_name} Train Acc')
    plt.plot(history1.history['val_accuracy'], label=f'{model1_name} Val Acc')
    plt.plot(history2.history['accuracy'], label=f'{model2_name} Train Acc', linestyle='--')
    plt.plot(history2.history['val_accuracy'], label=f'{model2_name} Val Acc', linestyle='--')
    plt.title('Model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(loc='lower right')

    # Plot Training Loss
    plt.subplot(1, 2, 2)
    plt.plot(history1.history['loss'], label=f'{model1_name} Train Loss')
    plt.plot(history1.history['val_loss'], label=f'{model1_name} Val Loss')
    plt.plot(history2.history['loss'], label=f'{model2_name} Train Loss', linestyle='--')
    plt.plot(history2.history['val_loss'], label=f'{model2_name} Val Loss', linestyle='--')
    plt.title('Model Loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(loc='upper right')

    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Adjust layout to make space for suptitle
    plt.show()


if __name__ == '__main__':
    # Ensure TensorFlow is using the GPU if available
    physical_devices = tf.config.list_physical_devices('GPU')
    if physical_devices:
        try:
            for device in physical_devices:
                tf.config.experimental.set_memory_growth(device, True)
            print(f"Using GPU: {physical_devices[0]}")
        except RuntimeError as e:
            # Memory growth must be set before GPUs have been initialized
            print(e)
    else:
        print("No GPU found, using CPU.")

    # Run the experiments
    experiment_results = run_experiments()

    # You can now use `experiment_results` to analyze data for your paper.
    # For example, print specific results:
    # print(experiment_results['MNIST']['Medium']['Custom_MLP_CustomGatedActivationLayer']['test_accuracy'])
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_classification, make_regression

# Custom Layer Implementation
class CustomDenseLayer(layers.Layer):
    def __init__(self, units, activation_step_relu=True, **kwargs):
        super(CustomDenseLayer, self).__init__(**kwargs)
        self.units = units
        self.activation_step_relu = activation_step_relu # To switch between standard ReLU and custom activation

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(input_shape[-1], self.units),
            initializer="glorot_uniform",
            trainable=True,
            name="kernel"
        )
        self.b = self.add_weight(
            shape=(self.units,), initializer="zeros", trainable=True, name="bias"
        )
        if self.activation_step_relu:
            self.w2 = self.add_weight(
                shape=(input_shape[-1], self.units), # Or self.units depending on desired interaction
                initializer="glorot_uniform",
                trainable=True,
                name="kernel2"
            )
            self.c = self.add_weight(
                shape=(self.units,), initializer="zeros", trainable=True, name="bias2"
            )
        super(CustomDenseLayer, self).build(input_shape)

    def call(self, inputs):
        linear_output = tf.matmul(inputs, self.w) + self.b
        if self.activation_step_relu:
            # Step function (Heaviside)
            step_output = tf.cast(linear_output > 0, dtype=tf.float32)

            # ReLU(w2x + c)
            linear_output2 = tf.matmul(inputs, self.w2) + self.c # Or use linear_output for w2?
                                                               # Based on ReLU(w2x+c), it should be inputs
            relu_output2 = tf.nn.relu(linear_output2)

            return step_output * relu_output2
        else:
            # Standard ReLU for comparison
            return tf.nn.relu(linear_output)

    def get_config(self):
        config = super(CustomDenseLayer, self).get_config()
        config.update({"units": self.units, "activation_step_relu": self.activation_step_relu})
        return config

# --- Helper Functions ---
def plot_history(history, title):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    metric_key = 'accuracy' if 'accuracy' in history.history else 'mae' if 'mae' in history.history else None
    val_metric_key = 'val_accuracy' if 'val_accuracy' in history.history else 'val_mae' if 'val_mae' in history.history else None

    if metric_key and val_metric_key:
        plt.subplot(1, 2, 2)
        plt.plot(history.history[metric_key], label=metric_key.capitalize())
        plt.plot(history.history[val_metric_key], label=f'Validation {metric_key.capitalize()}')
        plt.title(f'{title} - {metric_key.capitalize()}')
        plt.xlabel('Epoch')
        plt.ylabel(metric_key.capitalize())
        plt.legend()
    plt.tight_layout()
    plt.show()

# --- Benchmark 1: MNIST Classification ---
print("\n--- Starting MNIST Classification Benchmark ---")

# 1. Load and preprocess MNIST dataset
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = keras.datasets.mnist.load_data()
x_train_mnist = x_train_mnist.reshape(-1, 28 * 28).astype("float32") / 255.0
x_test_mnist = x_test_mnist.reshape(-1, 28 * 28).astype("float32") / 255.0
y_train_mnist = keras.utils.to_categorical(y_train_mnist, num_classes=10)
y_test_mnist = keras.utils.to_categorical(y_test_mnist, num_classes=10)

# 2. Define Models
input_shape_mnist = (28 * 28,)
num_classes_mnist = 10

# Baseline Model (Standard Dense with ReLU)
model_baseline_mnist = keras.Sequential(
    [
        keras.Input(shape=input_shape_mnist),
        layers.Dense(128, activation="relu", name="dense1_relu"),
        layers.Dense(64, activation="relu", name="dense2_relu"),
        layers.Dense(num_classes_mnist, activation="softmax", name="output"),
    ]
)

# Custom Layer Model
model_custom_mnist = keras.Sequential(
    [
        keras.Input(shape=input_shape_mnist),
        CustomDenseLayer(128, activation_step_relu=True, name="custom_dense1"),
        CustomDenseLayer(64, activation_step_relu=True, name="custom_dense2"),
        layers.Dense(num_classes_mnist, activation="softmax", name="output"), # Output layer typically softmax/linear
    ]
)

# Model with Custom Layer acting as standard ReLU (for apples-to-apples parameter count if w2,c are an issue)
# This model uses the CustomDenseLayer but sets activation_step_relu=False, so it just does ReLU(wx+b)
model_custom_relu_mnist = keras.Sequential(
    [
        keras.Input(shape=input_shape_mnist),
        CustomDenseLayer(128, activation_step_relu=False, name="custom_dense1_as_relu"),
        CustomDenseLayer(64, activation_step_relu=False, name="custom_dense2_as_relu"),
        layers.Dense(num_classes_mnist, activation="softmax", name="output"),
    ]
)


# 3. Compile Models
optimizer = keras.optimizers.Adam(learning_rate=1e-3)
loss_fn_mnist = "categorical_crossentropy"
metrics_mnist = ["accuracy"]

model_baseline_mnist.compile(optimizer=optimizer, loss=loss_fn_mnist, metrics=metrics_mnist)
model_custom_mnist.compile(optimizer=optimizer, loss=loss_fn_mnist, metrics=metrics_mnist)
model_custom_relu_mnist.compile(optimizer=optimizer, loss=loss_fn_mnist, metrics=metrics_mnist)

print("\nBaseline MNIST Model Summary:")
model_baseline_mnist.summary()
print("\nCustom Layer MNIST Model Summary:")
model_custom_mnist.summary()
print("\nCustom Layer (as ReLU) MNIST Model Summary:")
model_custom_relu_mnist.summary()


# 4. Train Models
epochs = 10 # Keep epochs low for quick testing, increase for paper
batch_size = 128
validation_split = 0.2

print("\nTraining Baseline MNIST Model...")
history_baseline_mnist = model_baseline_mnist.fit(
    x_train_mnist,
    y_train_mnist,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=validation_split,
    verbose=1
)

print("\nTraining Custom Layer MNIST Model...")
history_custom_mnist = model_custom_mnist.fit(
    x_train_mnist,
    y_train_mnist,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=validation_split,
    verbose=1
)

print("\nTraining Custom Layer (as ReLU) MNIST Model...")
history_custom_relu_mnist = model_custom_relu_mnist.fit(
    x_train_mnist,
    y_train_mnist,
    batch_size=batch_size,
    epochs=epochs,
    validation_split=validation_split,
    verbose=1
)

# 5. Evaluate Models
print("\n--- MNIST Evaluation ---")
loss_baseline_mnist, acc_baseline_mnist = model_baseline_mnist.evaluate(x_test_mnist, y_test_mnist, verbose=0)
print(f"Baseline MNIST - Test Loss: {loss_baseline_mnist:.4f}, Test Accuracy: {acc_baseline_mnist:.4f}")

loss_custom_mnist, acc_custom_mnist = model_custom_mnist.evaluate(x_test_mnist, y_test_mnist, verbose=0)
print(f"Custom Layer MNIST - Test Loss: {loss_custom_mnist:.4f}, Test Accuracy: {acc_custom_mnist:.4f}")

loss_custom_relu_mnist, acc_custom_relu_mnist = model_custom_relu_mnist.evaluate(x_test_mnist, y_test_mnist, verbose=0)
print(f"Custom Layer (as ReLU) MNIST - Test Loss: {loss_custom_relu_mnist:.4f}, Test Accuracy: {acc_custom_relu_mnist:.4f}")


# Plotting (optional, but good for papers)
plot_history(history_baseline_mnist, "Baseline MNIST Model")
plot_history(history_custom_mnist, "Custom Layer MNIST Model")
plot_history(history_custom_relu_mnist, "Custom Layer (as ReLU) MNIST Model")


# --- Benchmark 2: Synthetic Regression ---
print("\n--- Starting Synthetic Regression Benchmark ---")

# 1. Generate and preprocess synthetic regression data
X_reg, y_reg = make_regression(n_samples=2000, n_features=20, noise=0.1, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, test_size=0.2, random_state=42)

scaler_X_reg = StandardScaler()
X_train_reg_scaled = scaler_X_reg.fit_transform(X_train_reg)
X_test_reg_scaled = scaler_X_reg.transform(X_test_reg)

# Reshape y if it's not 2D ( Keras expects (batch_size, num_targets) )
if y_train_reg.ndim == 1:
    y_train_reg = y_train_reg.reshape(-1, 1)
if y_test_reg.ndim == 1:
    y_test_reg = y_test_reg.reshape(-1, 1)

# 2. Define Models
input_shape_reg = (X_train_reg_scaled.shape[1],)

# Baseline Model (Standard Dense with ReLU)
model_baseline_reg = keras.Sequential(
    [
        keras.Input(shape=input_shape_reg),
        layers.Dense(64, activation="relu", name="dense1_relu"),
        layers.Dense(32, activation="relu", name="dense2_relu"),
        layers.Dense(1, activation="linear", name="output"), # Linear activation for regression output
    ]
)

# Custom Layer Model
model_custom_reg = keras.Sequential(
    [
        keras.Input(shape=input_shape_reg),
        CustomDenseLayer(64, activation_step_relu=True, name="custom_dense1"),
        CustomDenseLayer(32, activation_step_relu=True, name="custom_dense2"),
        layers.Dense(1, activation="linear", name="output"),
    ]
)

# Model with Custom Layer acting as standard ReLU
model_custom_relu_reg = keras.Sequential(
    [
        keras.Input(shape=input_shape_reg),
        CustomDenseLayer(64, activation_step_relu=False, name="custom_dense1_as_relu"),
        CustomDenseLayer(32, activation_step_relu=False, name="custom_dense2_as_relu"),
        layers.Dense(1, activation="linear", name="output"),
    ]
)

# 3. Compile Models
optimizer_reg = keras.optimizers.Adam(learning_rate=1e-3)
loss_fn_reg = "mse" # Mean Squared Error for regression
metrics_reg = ["mae"] # Mean Absolute Error

model_baseline_reg.compile(optimizer=optimizer_reg, loss=loss_fn_reg, metrics=metrics_reg)
model_custom_reg.compile(optimizer=optimizer_reg, loss=loss_fn_reg, metrics=metrics_reg)
model_custom_relu_reg.compile(optimizer=optimizer_reg, loss=loss_fn_reg, metrics=metrics_reg)

print("\nBaseline Regression Model Summary:")
model_baseline_reg.summary()
print("\nCustom Layer Regression Model Summary:")
model_custom_reg.summary()
print("\nCustom Layer (as ReLU) Regression Model Summary:")
model_custom_relu_reg.summary()

# 4. Train Models
epochs_reg = 50 # Increase for better convergence
batch_size_reg = 32

print("\nTraining Baseline Regression Model...")
history_baseline_reg = model_baseline_reg.fit(
    X_train_reg_scaled,
    y_train_reg,
    batch_size=batch_size_reg,
    epochs=epochs_reg,
    validation_split=validation_split,
    verbose=1
)

print("\nTraining Custom Layer Regression Model...")
history_custom_reg = model_custom_reg.fit(
    X_train_reg_scaled,
    y_train_reg,
    batch_size=batch_size_reg,
    epochs=epochs_reg,
    validation_split=validation_split,
    verbose=1
)

print("\nTraining Custom Layer (as ReLU) Regression Model...")
history_custom_relu_reg = model_custom_relu_reg.fit(
    X_train_reg_scaled,
    y_train_reg,
    batch_size=batch_size_reg,
    epochs=epochs_reg,
    validation_split=validation_split,
    verbose=1
)

# 5. Evaluate Models
print("\n--- Regression Evaluation ---")
mse_baseline_reg, mae_baseline_reg = model_baseline_reg.evaluate(X_test_reg_scaled, y_test_reg, verbose=0)
print(f"Baseline Regression - Test MSE: {mse_baseline_reg:.4f}, Test MAE: {mae_baseline_reg:.4f}")

mse_custom_reg, mae_custom_reg = model_custom_reg.evaluate(X_test_reg_scaled, y_test_reg, verbose=0)
print(f"Custom Layer Regression - Test MSE: {mse_custom_reg:.4f}, Test MAE: {mae_custom_reg:.4f}")

mse_custom_relu_reg, mae_custom_relu_reg = model_custom_relu_reg.evaluate(X_test_reg_scaled, y_test_reg, verbose=0)
print(f"Custom Layer (as ReLU) Regression - Test MSE: {mse_custom_relu_reg:.4f}, Test MAE: {mae_custom_relu_reg:.4f}")

# Plotting
plot_history(history_baseline_reg, "Baseline Regression Model")
plot_history(history_custom_reg, "Custom Layer Regression Model")
plot_history(history_custom_relu_reg, "Custom Layer (as ReLU) Regression Model")


# --- Further Benchmarks to Consider for a Paper ---
# 1. Other Datasets:
#    - CIFAR-10/CIFAR-100: More complex image classification. You might need to combine your layer with Convolutional layers.
#    - UCI Datasets: Various classification/regression tasks (e.g., Iris, Wine, Boston Housing).
#    - IMDb Reviews: Text classification (requires text preprocessing and embedding layers).
#
# 2. Deeper Architectures: Test how your layer performs in deeper networks.
#
# 3. Different Optimizers and Learning Rates: Experiment with SGD, RMSprop, etc., and tune learning rates.
#
# 4. Regularization: Add L1/L2 regularization or Dropout to see its effect with your custom layer.
#
# 5. Parameter Efficiency: Compare the number of parameters of your custom model vs. a standard model
#    that achieves similar performance. Your custom layer has more parameters (w, b, w2, c) than a
#    standard Dense layer with ReLU (w, b). The `model_custom_relu_mnist/reg` is a control here.
#
# 6. Training Time / Convergence Speed: Record and compare wall-clock training times.
#
# 7. Ablation Studies:
#    - What if `w2` is not learned but fixed (e.g., tied to `w` or set to 1)?
#    - What if `c` is not learned but fixed (e.g., tied to `b` or set to 0)?
#    - Analyze the impact of the Step component vs. the learned ReLU component.
#
# 8. Visualization of Activations: Try to visualize the activations from your custom layer to understand
#    what it's learning.
#
# --- How to Structure Results for a Paper ---
#
# * **Introduction**: Briefly introduce the standard Dense layer and ReLU. Motivate your new layer.
# * **Proposed Method**: Clearly define your layer: `Step(Wx+b) * ReLU(w2x+c)`. Explain the role of `w, b, w2, c`.
# * **Experimental Setup**:
#     * **Datasets**: Describe the datasets used (MNIST, synthetic, etc.), including size, features, classes.
#     * **Baseline Models**: Describe the architecture of the standard models you're comparing against.
#     * **Proposed Models**: Describe the architecture of models using your custom layer.
#     * **Training Details**: Specify optimizer, learning rate, batch size, number of epochs, loss functions, evaluation metrics.
#     * **Hardware/Software**: Mention TensorFlow version, Python version, and optionally CPU/GPU details.
# * **Results**:
#     * Present tables comparing key metrics (accuracy, MSE, MAE, loss) on test sets.
#     * Include standard deviations if you run experiments multiple times with different random seeds.
#     * Show learning curves (loss/accuracy vs. epochs) for training and validation sets.
#     * Compare model sizes (number of parameters).
#     * Compare training times if relevant.
# * **Discussion/Analysis**:
#     * Interpret the results. Does your layer outperform the baselines? In what scenarios?
#     * Discuss any advantages (e.g., faster convergence, better generalization) or disadvantages (e.g., more parameters, slower computation per epoch).
#     * Analyze the learned parameters `w2`, `c`. Do they take on interesting values?
#     * Discuss limitations of your study and potential future work.
# * **Conclusion**: Summarize your findings.
#
# Remember to run each experiment multiple times (e.g., 3-5 times) with different random initializations
# and report the mean and standard deviation of the performance metrics for more robust results.
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.datasets import make_classification, make_regression, load_wine
import time # To record training times

# --- Custom Layer Implementation (from previous response) ---
class CustomDenseLayer(layers.Layer):
    def __init__(self, units, activation_step_relu=True, **kwargs):
        super(CustomDenseLayer, self).__init__(**kwargs)
        self.units = units
        self.activation_step_relu = activation_step_relu

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(input_shape[-1], self.units),
            initializer="glorot_uniform",
            trainable=True,
            name="kernel"
        )
        self.b = self.add_weight(
            shape=(self.units,), initializer="zeros", trainable=True, name="bias"
        )
        if self.activation_step_relu:
            self.w2 = self.add_weight(
                shape=(input_shape[-1], self.units),
                initializer="glorot_uniform",
                trainable=True,
                name="kernel2"
            )
            self.c = self.add_weight(
                shape=(self.units,), initializer="zeros", trainable=True, name="bias2"
            )
        super(CustomDenseLayer, self).build(input_shape)

    def call(self, inputs):
        linear_output = tf.matmul(inputs, self.w) + self.b
        if self.activation_step_relu:
            step_output = tf.cast(linear_output > 0, dtype=tf.float32)
            linear_output2 = tf.matmul(inputs, self.w2) + self.c
            relu_output2 = tf.nn.relu(linear_output2)
            return step_output * relu_output2
        else:
            return tf.nn.relu(linear_output)

    def get_config(self):
        config = super(CustomDenseLayer, self).get_config()
        config.update({"units": self.units, "activation_step_relu": self.activation_step_relu})
        return config

# --- Helper Functions ---
def plot_history(history, title, metric_name='accuracy', val_metric_name='val_accuracy'):
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()

    if metric_name in history.history and val_metric_name in history.history:
        plt.subplot(1, 2, 2)
        plt.plot(history.history[metric_name], label=metric_name.replace('_', ' ').title())
        plt.plot(history.history[val_metric_name], label=val_metric_name.replace('_', ' ').title())
        plt.title(f'{title} - {metric_name.replace("_", " ").title()}')
        plt.xlabel('Epoch')
        plt.ylabel(metric_name.replace('_', ' ').title())
        plt.legend()
    plt.tight_layout()
    plt.show()

def compile_and_train(model, x_train, y_train, x_val, y_val, optimizer, loss_fn, metrics, epochs, batch_size, model_name="Model"):
    model.compile(optimizer=optimizer, loss=loss_fn, metrics=metrics)
    print(f"\n--- {model_name} Summary ---")
    model.summary()
    print(f"\nTraining {model_name}...")
    start_time = time.time()
    history = model.fit(
        x_train,
        y_train,
        batch_size=batch_size,
        epochs=epochs,
        validation_data=(x_val, y_val),
        verbose=1
    )
    end_time = time.time()
    training_time = end_time - start_time
    print(f"{model_name} - Training Time: {training_time:.2f} seconds")
    return history, training_time

def evaluate_model(model, x_test, y_test, model_name="Model", training_time=None):
    loss, metric = model.evaluate(x_test, y_test, verbose=0)
    metric_name = model.metrics_names[1]
    print(f"{model_name} - Test Loss: {loss:.4f}, Test {metric_name.capitalize()}: {metric:.4f}")
    if training_time is not None:
        print(f"{model_name} - Training Time: {training_time:.2f} seconds")
    return loss, metric

# --- General Training Parameters ---
EPOCHS_GENERAL = 10 # Keep low for quick testing; increase for paper
BATCH_SIZE_GENERAL = 64
VALIDATION_SPLIT_GENERAL = 0.2 # Used if no separate val set is made
OPTIMIZER = keras.optimizers.Adam(learning_rate=1e-3)

# ==============================================================================
# Benchmark 1: MNIST Classification (from previous response, condensed)
# ==============================================================================
print("\n\n=== MNIST Classification Benchmark ===")
(x_train_mnist, y_train_mnist), (x_test_mnist, y_test_mnist) = keras.datasets.mnist.load_data()
x_train_mnist = x_train_mnist.reshape(-1, 28 * 28).astype("float32") / 255.0
x_test_mnist = x_test_mnist.reshape(-1, 28 * 28).astype("float32") / 255.0
y_train_mnist_cat = keras.utils.to_categorical(y_train_mnist, num_classes=10)
y_test_mnist_cat = keras.utils.to_categorical(y_test_mnist, num_classes=10)
x_train_mnist, x_val_mnist, y_train_mnist_cat, y_val_mnist_cat = train_test_split(
    x_train_mnist, y_train_mnist_cat, test_size=VALIDATION_SPLIT_GENERAL, random_state=42
)

input_shape_mnist = (28 * 28,)
num_classes_mnist = 10
loss_fn_mnist = "categorical_crossentropy"
metrics_mnist = ["accuracy"]

# Baseline MNIST
model_baseline_mnist = keras.Sequential([
    keras.Input(shape=input_shape_mnist),
    layers.Dense(128, activation="relu"), layers.Dense(64, activation="relu"),
    layers.Dense(num_classes_mnist, activation="softmax")], name="Baseline_MNIST")
history_baseline_mnist, time_baseline_mnist = compile_and_train(model_baseline_mnist, x_train_mnist, y_train_mnist_cat, x_val_mnist, y_val_mnist_cat, OPTIMIZER, loss_fn_mnist, metrics_mnist, EPOCHS_GENERAL, BATCH_SIZE_GENERAL, "Baseline MNIST")

# Custom MNIST
model_custom_mnist = keras.Sequential([
    keras.Input(shape=input_shape_mnist),
    CustomDenseLayer(128, activation_step_relu=True), CustomDenseLayer(64, activation_step_relu=True),
    layers.Dense(num_classes_mnist, activation="softmax")], name="Custom_MNIST")
history_custom_mnist, time_custom_mnist = compile_and_train(model_custom_mnist, x_train_mnist, y_train_mnist_cat, x_val_mnist, y_val_mnist_cat, OPTIMIZER, loss_fn_mnist, metrics_mnist, EPOCHS_GENERAL, BATCH_SIZE_GENERAL, "Custom MNIST")

# Custom as ReLU MNIST
model_custom_relu_mnist = keras.Sequential([
    keras.Input(shape=input_shape_mnist),
    CustomDenseLayer(128, activation_step_relu=False), CustomDenseLayer(64, activation_step_relu=False),
    layers.Dense(num_classes_mnist, activation="softmax")], name="Custom_ReLU_MNIST")
history_custom_relu_mnist, time_custom_relu_mnist = compile_and_train(model_custom_relu_mnist, x_train_mnist, y_train_mnist_cat, x_val_mnist, y_val_mnist_cat, OPTIMIZER, loss_fn_mnist, metrics_mnist, EPOCHS_GENERAL, BATCH_SIZE_GENERAL, "Custom (as ReLU) MNIST")

print("\n--- MNIST Evaluation ---")
evaluate_model(model_baseline_mnist, x_test_mnist, y_test_mnist_cat, "Baseline MNIST", time_baseline_mnist)
evaluate_model(model_custom_mnist, x_test_mnist, y_test_mnist_cat, "Custom MNIST", time_custom_mnist)
evaluate_model(model_custom_relu_mnist, x_test_mnist, y_test_mnist_cat, "Custom (as ReLU) MNIST", time_custom_relu_mnist)

plot_history(history_baseline_mnist, "Baseline MNIST Model")
plot_history(history_custom_mnist, "Custom Layer MNIST Model")
plot_history(history_custom_relu_mnist, "Custom Layer (as ReLU) MNIST Model")

# ==============================================================================
# Benchmark 2: CIFAR-100 Classification
# ==============================================================================
print("\n\n=== CIFAR-100 Classification Benchmark ===")
print("Note: For CIFAR-100, using purely Dense layers on flattened images is suboptimal.")
print("Consider using CNNs as a backbone for a more rigorous paper.")

(x_train_cifar100, y_train_cifar100), (x_test_cifar100, y_test_cifar100) = keras.datasets.cifar100.load_data()
x_train_cifar100 = x_train_cifar100.astype("float32") / 255.0
x_test_cifar100 = x_test_cifar100.astype("float32") / 255.0

# Flatten images for Dense layers
x_train_cifar100_flat = x_train_cifar100.reshape(-1, 32 * 32 * 3)
x_test_cifar100_flat = x_test_cifar100.reshape(-1, 32 * 32 * 3)

y_train_cifar100_cat = keras.utils.to_categorical(y_train_cifar100, num_classes=100)
y_test_cifar100_cat = keras.utils.to_categorical(y_test_cifar100, num_classes=100)

x_train_cifar100_flat, x_val_cifar100_flat, y_train_cifar100_cat, y_val_cifar100_cat = train_test_split(
    x_train_cifar100_flat, y_train_cifar100_cat, test_size=VALIDATION_SPLIT_GENERAL, random_state=42
)

input_shape_cifar100 = (32 * 32 * 3,)
num_classes_cifar100 = 100
loss_fn_cifar100 = "categorical_crossentropy"
metrics_cifar100 = ["accuracy"]
EPOCHS_CIFAR100 = 15 # CIFAR-100 is more complex, might need more epochs

# Baseline CIFAR-100 (Flattened)
model_baseline_cifar100 = keras.Sequential([
    keras.Input(shape=input_shape_cifar100),
    layers.Dense(512, activation="relu"), # Larger layer for more complex data
    layers.Dense(256, activation="relu"),
    layers.Dense(num_classes_cifar100, activation="softmax")], name="Baseline_CIFAR100_Flat")
history_baseline_cifar100, time_baseline_cifar100 = compile_and_train(model_baseline_cifar100, x_train_cifar100_flat, y_train_cifar100_cat, x_val_cifar100_flat, y_val_cifar100_cat, OPTIMIZER, loss_fn_cifar100, metrics_cifar100, EPOCHS_CIFAR100, BATCH_SIZE_GENERAL, "Baseline CIFAR-100 (Flattened)")

# Custom CIFAR-100 (Flattened)
model_custom_cifar100 = keras.Sequential([
    keras.Input(shape=input_shape_cifar100),
    CustomDenseLayer(512, activation_step_relu=True),
    CustomDenseLayer(256, activation_step_relu=True),
    layers.Dense(num_classes_cifar100, activation="softmax")], name="Custom_CIFAR100_Flat")
history_custom_cifar100, time_custom_cifar100 = compile_and_train(model_custom_cifar100, x_train_cifar100_flat, y_train_cifar100_cat, x_val_cifar100_flat, y_val_cifar100_cat, OPTIMIZER, loss_fn_cifar100, metrics_cifar100, EPOCHS_CIFAR100, BATCH_SIZE_GENERAL, "Custom CIFAR-100 (Flattened)")

# Custom as ReLU CIFAR-100 (Flattened)
model_custom_relu_cifar100 = keras.Sequential([
    keras.Input(shape=input_shape_cifar100),
    CustomDenseLayer(512, activation_step_relu=False),
    CustomDenseLayer(256, activation_step_relu=False),
    layers.Dense(num_classes_cifar100, activation="softmax")], name="Custom_ReLU_CIFAR100_Flat")
history_custom_relu_cifar100, time_custom_relu_cifar100 = compile_and_train(model_custom_relu_cifar100, x_train_cifar100_flat, y_train_cifar100_cat, x_val_cifar100_flat, y_val_cifar100_cat, OPTIMIZER, loss_fn_cifar100, metrics_cifar100, EPOCHS_CIFAR100, BATCH_SIZE_GENERAL, "Custom (as ReLU) CIFAR-100 (Flattened)")

print("\n--- CIFAR-100 (Flattened) Evaluation ---")
evaluate_model(model_baseline_cifar100, x_test_cifar100_flat, y_test_cifar100_cat, "Baseline CIFAR-100 (Flattened)", time_baseline_cifar100)
evaluate_model(model_custom_cifar100, x_test_cifar100_flat, y_test_cifar100_cat, "Custom CIFAR-100 (Flattened)", time_custom_cifar100)
evaluate_model(model_custom_relu_cifar100, x_test_cifar100_flat, y_test_cifar100_cat, "Custom (as ReLU) CIFAR-100 (Flattened)", time_custom_relu_cifar100)

plot_history(history_baseline_cifar100, "Baseline CIFAR-100 (Flattened) Model")
plot_history(history_custom_cifar100, "Custom Layer CIFAR-100 (Flattened) Model")
plot_history(history_custom_relu_cifar100, "Custom Layer (as ReLU) CIFAR-100 (Flattened) Model")

# --- Optional: CIFAR-100 with CNN Backbone (Conceptual Structure) ---
# print("\n--- Conceptual CIFAR-100 CNN Structure ---")
# def create_cnn_model_base(input_shape_cnn):
#     model = keras.Sequential([
#         keras.Input(shape=input_shape_cnn),
#         layers.Conv2D(32, (3,3), activation='relu', padding='same'),
#         layers.MaxPooling2D((2,2)),
#         layers.Conv2D(64, (3,3), activation='relu', padding='same'),
#         layers.MaxPooling2D((2,2)),
#         layers.Conv2D(128, (3,3), activation='relu', padding='same'),
#         layers.MaxPooling2D((2,2)),
#         layers.Flatten()
#     ])
#     return model

# cnn_base = create_cnn_model_base((32,32,3))
# model_baseline_cnn_cifar100 = keras.Sequential([cnn_base, layers.Dense(128, activation="relu"), layers.Dense(num_classes_cifar100, activation="softmax")])
# model_custom_cnn_cifar100 = keras.Sequential([cnn_base, CustomDenseLayer(128, activation_step_relu=True), layers.Dense(num_classes_cifar100, activation="softmax")])
# print("To run CNN versions, uncomment and adapt the training/evaluation loop for non-flattened x_train_cifar100 etc.")
# model_baseline_cnn_cifar100.compile(optimizer=OPTIMIZER, loss=loss_fn_cifar100, metrics=metrics_cifar100)
# model_baseline_cnn_cifar100.summary() # Just to show structure
# --- End Optional CNN ---


# ==============================================================================
# Benchmark 3: UCI Dataset (Wine) Classification
# ==============================================================================
print("\n\n=== UCI Wine Dataset Classification Benchmark ===")
wine_data = load_wine()
X_wine = wine_data.data
y_wine = wine_data.target

# Preprocess
scaler_wine = StandardScaler()
X_wine_scaled = scaler_wine.fit_transform(X_wine)

# One-hot encode labels
encoder_wine = OneHotEncoder(sparse_output=False) # Changed from sparse=False
y_wine_cat = encoder_wine.fit_transform(y_wine.reshape(-1, 1))

X_train_wine, X_test_wine, y_train_wine_cat, y_test_wine_cat = train_test_split(
    X_wine_scaled, y_wine_cat, test_size=0.2, random_state=42, stratify=y_wine # Stratify for class balance
)
X_train_wine, X_val_wine, y_train_wine_cat, y_val_wine_cat = train_test_split(
    X_train_wine, y_train_wine_cat, test_size=VALIDATION_SPLIT_GENERAL, random_state=42, stratify=y_train_wine_cat[:,0] # Stratify on first col of OHE y
)


input_shape_wine = (X_train_wine.shape[1],)
num_classes_wine = y_train_wine_cat.shape[1]
loss_fn_wine = "categorical_crossentropy"
metrics_wine = ["accuracy"]
EPOCHS_WINE = 30 # Small dataset, can train a bit more or risk overfitting quickly
BATCH_SIZE_WINE = 16 # Smaller batch for smaller dataset

# Baseline Wine
model_baseline_wine = keras.Sequential([
    keras.Input(shape=input_shape_wine),
    layers.Dense(64, activation="relu"),
    layers.Dense(32, activation="relu"),
    layers.Dense(num_classes_wine, activation="softmax")], name="Baseline_Wine")
history_baseline_wine, time_baseline_wine = compile_and_train(model_baseline_wine, X_train_wine, y_train_wine_cat, X_val_wine, y_val_wine_cat, OPTIMIZER, loss_fn_wine, metrics_wine, EPOCHS_WINE, BATCH_SIZE_WINE, "Baseline Wine")

# Custom Wine
model_custom_wine = keras.Sequential([
    keras.Input(shape=input_shape_wine),
    CustomDenseLayer(64, activation_step_relu=True),
    CustomDenseLayer(32, activation_step_relu=True),
    layers.Dense(num_classes_wine, activation="softmax")], name="Custom_Wine")
history_custom_wine, time_custom_wine = compile_and_train(model_custom_wine, X_train_wine, y_train_wine_cat, X_val_wine, y_val_wine_cat, OPTIMIZER, loss_fn_wine, metrics_wine, EPOCHS_WINE, BATCH_SIZE_WINE, "Custom Wine")

# Custom as ReLU Wine
model_custom_relu_wine = keras.Sequential([
    keras.Input(shape=input_shape_wine),
    CustomDenseLayer(64, activation_step_relu=False),
    CustomDenseLayer(32, activation_step_relu=False),
    layers.Dense(num_classes_wine, activation="softmax")], name="Custom_ReLU_Wine")
history_custom_relu_wine, time_custom_relu_wine = compile_and_train(model_custom_relu_wine, X_train_wine, y_train_wine_cat, X_val_wine, y_val_wine_cat, OPTIMIZER, loss_fn_wine, metrics_wine, EPOCHS_WINE, BATCH_SIZE_WINE, "Custom (as ReLU) Wine")

print("\n--- UCI Wine Evaluation ---")
evaluate_model(model_baseline_wine, X_test_wine, y_test_wine_cat, "Baseline Wine", time_baseline_wine)
evaluate_model(model_custom_wine, X_test_wine, y_test_wine_cat, "Custom Wine", time_custom_wine)
evaluate_model(model_custom_relu_wine, X_test_wine, y_test_wine_cat, "Custom (as ReLU) Wine", time_custom_relu_wine)

plot_history(history_baseline_wine, "Baseline Wine Model")
plot_history(history_custom_wine, "Custom Layer Wine Model")
plot_history(history_custom_relu_wine, "Custom Layer (as ReLU) Wine Model")

# ==============================================================================
# Benchmark 4: IMDB Movie Reviews Classification
# ==============================================================================
print("\n\n=== IMDB Movie Reviews Classification Benchmark ===")
MAX_FEATURES_IMDB = 10000  # Number of words to consider as features
MAXLEN_IMDB = 200        # Max sequence length
EMBEDDING_DIM_IMDB = 64

(x_train_imdb, y_train_imdb), (x_test_imdb, y_test_imdb) = keras.datasets.imdb.load_data(num_words=MAX_FEATURES_IMDB)

x_train_imdb_pad = keras.preprocessing.sequence.pad_sequences(x_train_imdb, maxlen=MAXLEN_IMDB)
x_test_imdb_pad = keras.preprocessing.sequence.pad_sequences(x_test_imdb, maxlen=MAXLEN_IMDB)

x_train_imdb_pad, x_val_imdb_pad, y_train_imdb, y_val_imdb = train_test_split(
    x_train_imdb_pad, y_train_imdb, test_size=VALIDATION_SPLIT_GENERAL, random_state=42
)


loss_fn_imdb = "binary_crossentropy" # Binary classification
metrics_imdb = ["accuracy"]
EPOCHS_IMDB = 5 # Text data can train fast but also overfit quickly
BATCH_SIZE_IMDB = 128

# Baseline IMDB
model_baseline_imdb = keras.Sequential([
    keras.Input(shape=(MAXLEN_IMDB,)),
    layers.Embedding(MAX_FEATURES_IMDB, EMBEDDING_DIM_IMDB),
    layers.GlobalAveragePooling1D(), # Or layers.Flatten()
    layers.Dense(64, activation="relu"),
    layers.Dense(1, activation="sigmoid")], name="Baseline_IMDB") # Sigmoid for binary output
history_baseline_imdb, time_baseline_imdb = compile_and_train(model_baseline_imdb, x_train_imdb_pad, y_train_imdb, x_val_imdb_pad, y_val_imdb, OPTIMIZER, loss_fn_imdb, metrics_imdb, EPOCHS_IMDB, BATCH_SIZE_IMDB, "Baseline IMDB")

# Custom IMDB
model_custom_imdb = keras.Sequential([
    keras.Input(shape=(MAXLEN_IMDB,)),
    layers.Embedding(MAX_FEATURES_IMDB, EMBEDDING_DIM_IMDB),
    layers.GlobalAveragePooling1D(),
    CustomDenseLayer(64, activation_step_relu=True),
    layers.Dense(1, activation="sigmoid")], name="Custom_IMDB")
history_custom_imdb, time_custom_imdb = compile_and_train(model_custom_imdb, x_train_imdb_pad, y_train_imdb, x_val_imdb_pad, y_val_imdb, OPTIMIZER, loss_fn_imdb, metrics_imdb, EPOCHS_IMDB, BATCH_SIZE_IMDB, "Custom IMDB")

# Custom as ReLU IMDB
model_custom_relu_imdb = keras.Sequential([
    keras.Input(shape=(MAXLEN_IMDB,)),
    layers.Embedding(MAX_FEATURES_IMDB, EMBEDDING_DIM_IMDB),
    layers.GlobalAveragePooling1D(),
    CustomDenseLayer(64, activation_step_relu=False),
    layers.Dense(1, activation="sigmoid")], name="Custom_ReLU_IMDB")
history_custom_relu_imdb, time_custom_relu_imdb = compile_and_train(model_custom_relu_imdb, x_train_imdb_pad, y_train_imdb, x_val_imdb_pad, y_val_imdb, OPTIMIZER, loss_fn_imdb, metrics_imdb, EPOCHS_IMDB, BATCH_SIZE_IMDB, "Custom (as ReLU) IMDB")

print("\n--- IMDB Evaluation ---")
evaluate_model(model_baseline_imdb, x_test_imdb_pad, y_test_imdb, "Baseline IMDB", time_baseline_imdb)
evaluate_model(model_custom_imdb, x_test_imdb_pad, y_test_imdb, "Custom IMDB", time_custom_imdb)
evaluate_model(model_custom_relu_imdb, x_test_imdb_pad, y_test_imdb, "Custom (as ReLU) IMDB", time_custom_relu_imdb)

plot_history(history_baseline_imdb, "Baseline IMDB Model")
plot_history(history_custom_imdb, "Custom Layer IMDB Model")
plot_history(history_custom_relu_imdb, "Custom Layer (as ReLU) IMDB Model")

print("\n\n=== Experimentation Complete ===")
print("Reminder: For a paper, increase epochs, run multiple seeds, perform hyperparameter tuning,")
print("and consider more complex architectures where appropriate (e.g., CNNs for CIFAR).")
print("Also, analyze parameter counts and training times carefully.")
```
