```python
import numpy as np
import matplotlib.pyplot as plt

def get_positional_encoding(max_seq_len, d_model):
    position = np.arange(max_seq_len)[:, np.newaxis]
    div_term = np.power(10000.0, np.arange(0, d_model, 2) / d_model)

    pos_encoding = np.zeros((max_seq_len, d_model))
    pos_encoding[:, 0::2] = np.sin(position / div_term)
    pos_encoding[:, 1::2] = np.cos(position / div_term)
    
    return pos_encoding

# Settings
MAX_SEQUENCE_LENGTH = 10000
EMBEDDING_DIM = 1024

pe = get_positional_encoding(MAX_SEQUENCE_LENGTH, EMBEDDING_DIM)

# Plot 1: Positional Embedding Heatmap
plt.figure(figsize=(10, 8))
plt.imshow(pe, cmap='viridis', aspect='auto')
plt.title('Sinusoidal Positional Embeddings')
plt.xlabel('Embedding Dimension')
plt.ylabel('Token Position')
plt.colorbar()
plt.show()

# Plot 2: Dot Product of Position Embeddings
dot_products = np.dot(pe, pe.T)
plt.figure(figsize=(10, 8))
plt.imshow(dot_products, cmap='plasma', aspect='auto')
plt.title('Dot Product Similarity Between Positions')
plt.xlabel('Token Position')
plt.ylabel('Token Position')
plt.colorbar()
plt.show()
```
